```python
#Replace()- In this method, we can add or replace some of the value of the dataframe by some defined external values.

initial_data = {'First_name': ['Ram', 'Mohan', 'Tina', 'Jeetu', 'Meera'], 
                'Last_name': ['Kumar', 'Sharma', 'Ali', 'Gandhi', 'Kumari'], 
                'Age': [42, 52, 36, 21, 23], 
                'City': ['Mumbai', 'Noida', 'Pune', 'Delhi', 'Bihar']}
df = pd.DataFrame(initial_data)
```

```python
df
```

```python
new_data = { "Ram":"Shyam",
             "Tina":"Riya",
             "Jeetu":"Jitender" }
# combine this new data with existing DataFrame
df = df.replace({'First_name':new_data}) # remeber while replace we need to pass the data in dict format
```

```python
df
```
