```python
import pandas as pd
import numpy as np
```

```python
column=['a','b','c','d','e']
index=['A','B','C','D','E']

df = pd.DataFrame(np.random.rand(5,5), columns = column, index =index)
print('original',df)
df.reindex((['B', 'D', 'A', 'C', 'E'])) # here we just chaning the sequence
```

```python
new_index = ['i','B','A','U','E']
print(df.reindex(new_index)) # the index which doesnt match with the original index will result in NaN
```
